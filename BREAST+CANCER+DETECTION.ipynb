```python
#LOADING THE DATASET FROM KAGGLE LINK : https://www.kaggle.com/rohitjain2086/breast-cancer-dataset-prediction/data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


df=pd.read_csv('/home/sayantan/Desktop/data.csv')

df.tail()
```

```python
#DROPPING THE NAN COLUMN
df.drop(['Unnamed: 32'],axis=1,inplace=True)
df.head()
```

```python
#CONVERTING THE OUTCOME(DIAGNOSIS) ONTO 0'S AND 1'S : 1 FOR MALIGNANT 0 FOR BENINE
df.diagnosis.replace('M',1,inplace=True)
df.diagnosis.replace('B',0,inplace=True)
```

```python
#CHECKING VALUES HAVE BEEN UPDATED
df.head()
```

```python
#DROPPING THE ID COLUMN WHICH IS OF NO USE
df.drop(['id'],axis=1,inplace=True)
df.head()
```

```python
#NOW WE WOULD FIND THE CO-RELATION LETS SEE
import seaborn as sns
features=df.columns[1:] #LEAVING THE DIAGNOSIS AS IT'S THE OUTCOME 
f=dict()
c1=list()
for i in features:
    f[str(i)]=df['diagnosis'].corr(df[str(i)])
    c1.append(df['diagnosis'].corr(df[str(i)]))
print(f.items())
#SO BASICALLY I CREATED A DICTIONARY AND MAPPED FEATURES WITH THEIR CO-RELATION
```

```python
#NOW WE FIND KEY FEATURES BASED ON THE CO-RELATION

z=sorted(f.values(),reverse=True)
z
#HERE WE SEE MOST FEATURES ARE HIGHLY CO-RELATED EXCEPT 3 OR 4 FEATURES HENCE WE WOULD NOT CONSIDER THEM
```

```python
#MODIFYING OUR FEATURES NOW
zmod=list()
for i in z:
    if i>0.2:
        zmod.append(i)
feat=list()
for j in f.keys():
    for k in zmod:
        if f[str(j)]==k:
            feat.append(str(j))
print("FEATURES SELECTED :\n")
for i in range(0,len(feat)):
    print(feat[i])
```

```python
#NOW WE DESCRIBE OUR DATASET
df[feat].describe()
```

```python
#DEFINING A FUNCTION TO REMOVE OUTLIERS
def outliers(name):
    q1,q2,q3=df[str(name)].quantile([0.25,0.5,0.75])
    iqr=q3-q1
    ll=q1-(1.5*(iqr))
    ul=q3+(1.5*(iqr))
    c=0
    for i in df[str(name)]:
        if i<ll or i>ul:
            c=c+1
            if(c==1):
                print("REMOVING THE OUTLIER.........")
            df.drop(df[df[str(name)]==i].index,inplace=True)
```

```python
#REMOVING OUTLIERS FROM OUR DATA
for i in feat:
    outliers(i)
```

```python
import seaborn as sns
df[feat].describe()
#COMPARE THIS RESULT TO THE PREVIOUS ONE THE OUTLIER'S HAVE BEEN REMOVED
```

```python
#DIVIDING INTO TRAIN AND TEST DATASETS
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.4)
test[test.diagnosis==1]
```

```python
#BUILDING RANDOM FOREST CLASSIFIER TO PREDICT THE OUTCOME
from sklearn.ensemble import RandomForestClassifier as rfc
from sklearn.metrics import accuracy_score
c1=rfc(n_jobs=2,n_estimators=200,criterion='gini')
c1.fit(train[feat],train['diagnosis'])
k=c1.predict(test[feat])
print(accuracy_score(k,test['diagnosis'])*100)
```

```python
#BUILDING THE LOGISTIC REGRESSION CLASSIFIER TO PREDICT AND USING GRIDSEARCHCV TO GET OPTIMAL VALUE OF 'C'
from sklearn.linear_model import LogisticRegression as lrc
from sklearn.model_selection import GridSearchCV
c2=lrc()
par_grid=[{'C':[0.001,0.01,0.1,1]}]
gs=GridSearchCV(estimator=c2,param_grid=par_grid,cv=10)

gs.fit(train[feat],train.diagnosis)

print("effective inverse regularization parameter : ",gs.best_estimator_.get_params()['C'])
n=gs.best_estimator_.get_params()['C']


c3=lrc(C=n)
c3.fit(train[feat],train['diagnosis'])
k1=c3.predict(test[feat])
print(accuracy_score(k1,test['diagnosis'])*100)
```

```python
#GAUSSIAN NAIVE-BAYES CLASSIFIER
from sklearn.naive_bayes import GaussianNB as gn
pos=df[df.diagnosis==1].diagnosis.count()
neg=df[df.diagnosis==0].diagnosis.count()
tot=df.diagnosis.count()
pos1=pos/tot
pos2=neg/tot
p=np.array([[pos1],[pos2]])

c4=gn(p)
c4.fit(train[feat],train.diagnosis)
k2=c4.predict(test[feat])
print(accuracy_score(k2,test.diagnosis)*100)
```
